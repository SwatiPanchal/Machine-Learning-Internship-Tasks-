```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from statistics import mean,stdev
from sklearn.model_selection import StratifiedKFold
```

```python
cancer = load_breast_cancer()
```

```python
cancer
```

```python
X = cancer.data
```

```python
y = cancer.target
```

```python
X.shape, y.shape
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(X)
```

```python
lr = LogisticRegression()
```

```python
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)
accu_stratified = []
```

```python
for i in skf.split(X,y):
    print(i)
```

```python
c = 0
for train_index, test_index in skf.split(X,y):
    print(train_index)
    x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    lr.fit(x_train_fold,y_train_fold)
    accu_stratified.append(lr.score(x_test_fold,y_test_fold))
```

```python
print('List of possible accuracy:', accu_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(accu_stratified)*100, '%')
print('\nMinimum Accuracy:',
      min(accu_stratified)*100, '%')
print('\nOverall Accuracy:',
      mean(accu_stratified)*100, '%')
print('\nStandard Deviation is:', stdev(accu_stratified))
```

### GridSearchCV

```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV

data = load_breast_cancer()
 
X = data.data
y = data.target

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=42)
svc = SVC()
svc.fit(Xtrain, ytrain)

pred = svc.predict(Xtest)
print(confusion_matrix(ytest,pred))
print(classification_report(ytest,pred))
```

```python

param_grid = {'C': [0.1, 1, 10, 100],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'gamma':['scale', 'auto'],
              'kernel': ['linear']}  

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3, n_jobs=-1)

grid.fit(Xtrain, ytrain)

print(grid.best_params_)

grid_pred = grid.predict(Xtest)

print(confusion_matrix(ytest,grid_pred))
print(classification_report(ytest,grid_pred))
```
